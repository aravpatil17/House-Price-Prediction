## DATA PREPARATION

```python
varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(x):
    return x.map({'yes': 1, "no": 0})
df[varlist] = df[varlist].apply(binary_map)
```

```python
df.head()
```

We need dummy variables for furnishing status to convert the values to integers or boolean values.

```python
status = pd.get_dummies(df['furnishingstatus'])
```

```python
status.head()
```

Now, you don't need three columns. You can drop the furnished column, as the type of furnishing can be identified with just the last two columns where —

1) 00 will correspond to furnished
2) 01 will correspond to unfurnished
3) 10 will correspond to semi-furnished

```python
status = pd.get_dummies(df['furnishingstatus'], drop_first = True)
```

```python
df = pd.concat([df, status], axis = 1)
```

```python
df.head()
```

We now drop furnishing status as we have created dummies for it.

```python
df.drop(['furnishingstatus'], axis = 1, inplace = True)
```

```python
df.head()
```

## SPLITTING THE DATA INTO TRAINING AND TESTING SETS

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

## RESCALING THE FEATURES

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

The above graph was to check the correlation coefficients to see which variables are highly correlated.
As we can see, area seems to the correlated to price the most.

```python
y_train = df_train.pop('price')
X_train = df_train
```

## BUILDING OUR MODEL

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
rfe = RFE(lm, n_features_to_select = 6)            
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

```python
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## RESIDUAL ANALYSIS OF THE TRAIN DATA

```python
y_train_price = lm.predict(X_train_rfe)
```

```python
res = (y_train_price - y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)               
plt.xlabel('Errors', fontsize = 18)  
```

```python
plt.scatter(y_train,res)
plt.show()
```
